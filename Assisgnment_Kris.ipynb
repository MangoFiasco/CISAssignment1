```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

housingInfo = pd.read_csv('housing/train.csv')
housingInfo = housingInfo.sort_values("SalePrice")

#Data to Analyze from main data############
lotArea = housingInfo['LotArea']
overallQ = housingInfo['OverallQual']
overallC = housingInfo['OverallCond']
yearB = housingInfo['YearBuilt']
yearR = housingInfo['YearRemodAdd']
############################################

#Price
price = housingInfo['SalePrice']

```

```python
#OverallQual Vs Price Model

overallQVPrice = pd.DataFrame({"OverallQual":overallQ,"SalePrice":price})
X = overallQ
Y = price
n = overallQVPrice.shape[0]
plt.figure(figsize=(10,10))
plt.scatter(X,Y,3)
plt.xlim(0,20)
w = 20000.
b = 300.
x = np.array(range(0,20,1))
y = (w*x + b)
plt.plot(x,y)
plt.show()
```

```python
#LOSS 
def getLoss(X,y,w,b):
    i = 0
    MSE = 0.
    elements = X.shape[0]
    while i<overallQVPrice.shape[0]:
        MSE = MSE + (float(y[i]) - (w*float(X[i])+b))**2
        i = i+1
    MSE = MSE/i
    return MSE

print(getLoss(X,Y,w,b))
```

```python
# GRADIENT DECENT!
desc = 0
w = 0
b = 0
while desc < 100:
    learningRate = 1000
    yGuess = w * X + b
    wGrd = -(2/n) * sum(X * (Y - yGuess))
    bGrd =  -(2/n) * sum(Y - yGuess)

    newW = w - wGrd/learningRate
    newB = b - bGrd/learningRate
    print(getLoss(X,Y,newW,newB))
    w = newW
    b = newB
    desc = desc + 1
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X,Y,1)
plt.xlim(0,20)
plt.ylim(0,800000)
x = np.array(range(0,20,1))
y = (newW*x + newB)
plt.plot(x,y)
plt.show()
#print(newW, newB)
```
